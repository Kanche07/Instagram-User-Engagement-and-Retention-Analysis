## **Instagram: Improving Engagement Quality & Retention.**

**Business Context**

Instagram is one of the world’s largest social platforms, used daily for entertainment, discovery, and communication across formats like Feed (photos), Reels, Stories, Explore, and DMs.

As Instagram scales, leadership’s focus is shifting from “more time spent” to Engagement Quality — making sure that user activity is healthy, consistent, and sustainable, rather than driven by unhealthy patterns (doom-scrolling), notification spam, or low-quality interaction (passive views with no meaningful actions).

Poor engagement quality creates:

lower long-term retention (users churn after heavy short-term usage),
weaker creator ecosystem (less meaningful engagement),
higher trust/safety risk (account security issues, privacy distrust),
weaker monetization (premium users don’t stick, ad engagement declines).
Instagram believes the answers lie in its user lifestyle + usage behavior dataset, capturing engagement patterns, content preferences, notification behavior, and account settings.

# **Objective:**

To identify and improve healthy, meaningful user engagement on Instagram by analyzing Engagement Quality Efficiency (EQE), and to reduce churn risk by discouraging passive doom-scrolling behaviors while improving long-term retention and user well-being.

In simple **"Instagram wants users to engage meaningfully, not just scroll endlessly."**

More Specifically the objective is:
*   Healthy engagement
*   long-term retention
*   Meaningful interactions
*   Reduced churn and dormancy risk.

**A) Deconstruct EQE across key dimensions**
*   demographics: age, gender
*   platform behavior: sessions_per_day
*   content: content_type_perfernce, preferred_content_theme
*   account trust: privacy_setting_level

**B) Identify which combinations drive**
*   high EQE(healthy engagement)
*   low EQE (passive doom-scroll patterns)
*   dormancy risk(low sessions + low notif response + old last_login_date)

**c) Study Lifestyle links**
check if health/lifestyle correlate with engagement quality:

*   sleep_hours_per_night, exercise_hours_per_week, perceived_stress_score
*   and see association with key metrics

**D) Surface Actionable insights**
Recommend actions to:
*   Improve feed ranking for meaningful engagement
*   reduce passive scrolling better
*   personalize notifications better
*   improve retention for at-risk cohorts
*   increase premium  adoption(only if it improves EQE and retention)

# **Key metrics calculation**

**Calculating Attention(A)**

```python
attention_columns = ['time_on_feed_per_day','time_on_reels_per_day','time_on_explore_per_day','time_on_messages_per_day']
instagram_df['attention'] = instagram_df[attention_columns].sum(axis=1)
instagram_df[['user_id']+attention_columns+['attention']].head()
```

**Calculating Meaningful Engagaement.**
Meaningful Engagement is nothing but intent and effort behind user interactions on instagram.Key Engagement actions such as likes, comments, Direct Messages and posts.Each weights are assigned accordingly.

**like** --> less effort --> weight = 1

**comments** --> moderate effect --> weight = 3

**direct** **messages**(DM) --> high intent --> weight = 4

**posts** --> high effort --> weight = 5

**shares/saves** --> column doesn't exist in dataset

To maintain consistency across metrics, weekly actions (DMs and posts) are converted to daily equivalents by **dividing their values by 7**.



```python
instagram_df['dms_sent_per_day'] = instagram_df['dms_sent_per_week'] / 7
instagram_df['posts_created_per_day'] = instagram_df['posts_created_per_week'] / 7

#computing ME with weights
instagram_df['ME'] = (1 * instagram_df['likes_given_per_day']) + (3 * instagram_df['comments_written_per_day']) + (4 * instagram_df['dms_sent_per_day']) + (5 * instagram_df['posts_created_per_day'])
instagram_df[['user_id','likes_given_per_day','comments_written_per_day','dms_sent_per_day','posts_created_per_day','ME']].head()
```

**Calculating Engagement Quality Efficiency(EQE)**

```python
instagram_df['EQE'] = np.where(instagram_df['attention'] > 0,  instagram_df['ME'] / instagram_df['attention'], np.nan)
instagram_df[['user_id','attention','ME','EQE']].head()
```

**Statistical summary**

```python
instagram_df[['attention','ME','EQE']].describe(percentiles=[0.25,0.5,0.75]).T
```

**OBSERVATION:**
1. ME, Attention, EQE are rightly-skewed(mean>median)

2. Instagram engagement metrics was right-skewed because a few active users have much more activity than others.Which makes median more better measure than mean.

**Calculating High_EQE and Low_EQE categories**

```python
EQE_q25 = instagram_df['EQE'].quantile(0.25)
EQE_q75 = instagram_df['EQE'].quantile(0.75)

instagram_df['EQE_CATEGORY']=np.where(instagram_df['EQE']<= EQE_q25,'Low_EQE',np.where(instagram_df['EQE']>= EQE_q75,'High_EQE',None))
instagram_df['EQE_CATEGORY'].value_counts().reset_index(name='count')
```

```python
#segment wise comparison using median Attention(A), ME,EQE
instagram_df.groupby('EQE_CATEGORY')[['attention','ME','EQE']].median()
```

**Observation:**
HIGH_EQE ---> User spends less time with meaningful interactions.
LOW_EQE ---> User spends more time with less efficiency.

**Supporting Metrics**

1.Interaction Health
*   Comment-to-like-ratio
*   DM Acitivity Index



```python
instagram_df['comment_to_like_ratio'] = instagram_df['comments_written_per_day'] / instagram_df['likes_given_per_day']
instagram_df['dm_activity_index'] = (instagram_df['dms_sent_per_week'] + instagram_df['dms_received_per_week'])/7
supporting_metrics=['comment_to_like_ratio','dm_activity_index']
instagram_df.groupby('EQE_CATEGORY')[supporting_metrics].median()
```

2. Ads Efficiency
*   Ad CTR Proxy
*   Ad Load Proxy



```python
instagram_df['ad_CTR_proxy'] = instagram_df['ads_clicked_per_day'] / instagram_df['ads_viewed_per_day']
instagram_df['ad_load_proxy'] = instagram_df['ads_viewed_per_day'] / instagram_df['attention']
supporting_metrics=['ad_CTR_proxy','ad_load_proxy']
instagram_df.groupby('EQE_CATEGORY')[supporting_metrics].median()
```

2. Low dormancy risk users mostly show low EQE, meaning they log in frequently but mainly scroll without meaningful interaction.

**RECOMMENDATION:**
1. Improve EQE for low dormancy risk users by encouraging actions like comments, DMs. saves and shares instead of just increasing time spent.
